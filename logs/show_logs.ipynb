```python
import pandas as pd
import os

# Load the CSV file
log_dir = "./logs/"  # Replace with your actual log directory
csv_file = log_dir + "std_2Mio.csv"  # Replace with the actual name of your CSV file if different

# Read the CSV file into a pandas DataFrame
df = pd.read_csv(csv_file)

print(df.head(-10))
```

```python
# Plotten der Daten
import matplotlib.pyplot as plt

# Create a scatter plot for rollout/ep_rew_mean per iteration
plt.figure(figsize=(10, 6))
plt.scatter(df["time/total_timesteps"], df["rollout/ep_rew_mean"], color='blue')
plt.title("Episode Reward Mean per Timestamp")
plt.xlabel("Total timesteps")
plt.ylabel("Episode Reward Mean")
plt.grid(True)
plt.show()
```

```python
# Definiere die Spalten, die nicht geplottet werden sollen
columns_to_exclude = ['time/time_elapsed', 'time/total_timesteps', 'time/fps', "time/iterations", "train/clip_range", "train/clip_fraction", "train/n_updates", "train/learning_rate", "rollout/ep_len_mean"]  # Füge hier die Spalten hinzu, die du ausschließen möchtest

# Definiere die Spalten, die geplottet werden sollen
columns_to_plot = [column for column in df.columns if column not in columns_to_exclude]

# Plotte jede Metrik als Scatterplot
for column in columns_to_plot:
    plt.figure(figsize=(10, 6))
    plt.scatter(df['time/total_timesteps'], df[column], alpha=0.75, color='blue')
    plt.title(f'{column} over Time')
    plt.xlabel('Total timestamps')
    plt.ylabel(column)
    plt.grid(True)
    plt.show()
```
