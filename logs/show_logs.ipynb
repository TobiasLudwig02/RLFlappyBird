```python
import pandas as pd

# Load the CSV files
path_PPO_MLP_std_2Mio = "PPO_MLP_std_2Mio/progress.csv" 
path_PPO_MLP_rew100_2Mio = "PPO_MLP_rew100_2Mio/progress.csv" 
path_baseline_std_2Mio = "baseline_std_2Mio/progress.csv" 
path_DQN_MLP_std_2Mio = "DQN_MLP_std_2Mio/progress.csv" 
path_DQN_MLP_rew100_2Mio = "DQN_MLP_rew100_2Mio/progress.csv" 
path_A2C_MLP_std_2Mio= "A2C_MLP_std_2Mio/progress.csv" 
path_A2C_MLP_rew100_2Mio= "A2C_MLP_rew100_2Mio/progress.csv" 
path_curr_learning = "PPO_MLP_cur_learning/progress.csv"

# Read the CSV files into a pandas DataFrame
df_PPO_MLP_std_2Mio = pd.read_csv(path_PPO_MLP_std_2Mio)
df_PPO_MLP_rew100_2Mio = pd.read_csv(path_PPO_MLP_rew100_2Mio)
df_baseline_std_2Mio = pd.read_csv(path_baseline_std_2Mio)
df_DQN_MLP_std_2Mio = pd.read_csv(path_DQN_MLP_std_2Mio)
df_DQN_MLP_rew100_2Mio = pd.read_csv(path_DQN_MLP_rew100_2Mio)
df_A2C_MLP_std_2Mio = pd.read_csv(path_A2C_MLP_std_2Mio)
df_A2C_MLP_rew100_2Mio = pd.read_csv(path_A2C_MLP_rew100_2Mio)
df_cur_learning = pd.read_csv(path_curr_learning)
```

```python
import matplotlib.pyplot as plt

# Configure style
plt.style.use('seaborn-v0_8-darkgrid')  

colors = {
    'PPO': '#408AAD',  # blue
    'baseline': '#FF5733',  # red
    'DQN': '#9370DB',  # purple
    'A2C': '#E9967A',  # orange
    'Cur': '#228B22'   # green
}
s = 5
alpha=0.5
```

```python
# Plot reward from env with improved reward with 2 Mio timestamps

plt.figure(figsize=(10, 6))
plt.scatter(df_PPO_MLP_rew100_2Mio['time/total_timesteps'], df_PPO_MLP_rew100_2Mio["rollout/ep_rew_mean"], alpha=alpha, label='PPO_MLP_rew100_2Mio', color=colors['PPO'], s=s)
plt.scatter(df_DQN_MLP_rew100_2Mio['time/total_timesteps'], df_DQN_MLP_rew100_2Mio["rollout/ep_rew_mean"], alpha=alpha, label='DQN_MLP_rew100_2Mio', color=colors['DQN'], s=s)
plt.scatter(df_A2C_MLP_rew100_2Mio['time/total_timesteps'], df_A2C_MLP_rew100_2Mio["rollout/ep_rew_mean"], alpha=alpha, label='A2C_MLP_rew100_2Mio', color=colors['A2C'], s=s)
plt.xlabel('Total timestamps')
plt.ylabel("Mean reward per episode")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot reward from std env with 2 Mio timestamps

plt.figure(figsize=(10, 6))
plt.scatter(df_PPO_MLP_std_2Mio['time/total_timesteps'], df_PPO_MLP_std_2Mio["rollout/ep_rew_mean"], alpha=alpha, label='PPO_MLP_std_2Mio', color=colors['PPO'], s=s)
plt.scatter(df_DQN_MLP_std_2Mio['time/total_timesteps'], df_DQN_MLP_std_2Mio["rollout/ep_rew_mean"], alpha=alpha, label='DQN_MLP_std_2Mio', color=colors['DQN'], s=s)
plt.scatter(df_A2C_MLP_std_2Mio['time/total_timesteps'], df_A2C_MLP_std_2Mio["rollout/ep_rew_mean"], alpha=alpha, label='A2C_MLP_std_2Mio', color=colors['A2C'], s=s)
plt.scatter(df_baseline_std_2Mio['time/total_timesteps'], df_baseline_std_2Mio["rollout/ep_rew_mean"], alpha=alpha, label='Baseline_std_2Mio', color=colors['baseline'], s=s)
plt.xlabel('Total timestamps')
plt.ylabel("Mean reward per episode")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot reward from curriculum learning and std ppo

plt.figure(figsize=(10, 6))
plt.scatter(df_PPO_MLP_std_2Mio['time/total_timesteps'], df_PPO_MLP_std_2Mio["rollout/ep_rew_mean"], alpha=alpha, label='PPO_MLP_std_2Mio', color=colors['PPO'], s=s)
plt.scatter(df_cur_learning['time/total_timesteps'], df_cur_learning["rollout/ep_rew_mean"], alpha=alpha, label='PPO_MLP_std_Cur_learning', color=colors['Cur'], s=s)
plt.xlabel('Total timestamps')
plt.ylabel("Mean reward per episode")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Columns not to plot
columns_to_exclude = ['time/time_elapsed', 'time/total_timesteps', 'time/fps', "time/iterations", "train/clip_range", "train/clip_fraction", "train/n_updates", "train/learning_rate", "rollout/ep_len_mean"]  # Füge hier die Spalten hinzu, die du ausschließen möchtest

# Columns to plot
columns_to_plot = [column for column in df_PPO_MLP_std_2Mio.columns if column not in columns_to_exclude]

for column in columns_to_plot:
    plt.figure(figsize=(10, 6))
    plt.scatter(df_PPO_MLP_std_2Mio['time/total_timesteps'], df_PPO_MLP_std_2Mio[column], alpha=alpha, label='df_PPO_MLP_std_2Mio', color='blue', s=s)
    plt.scatter(df_cur_learning['time/total_timesteps'], df_cur_learning[column], alpha=alpha, label='df_cur_learning', color='green', s=s)
    plt.title(f'{column} over Time')
    plt.xlabel('Total timestamps')
    plt.ylabel(column)
    plt.legend()
    plt.grid(True)
    plt.show()
```
