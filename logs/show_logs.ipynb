```python
import pandas as pd

# Load the CSV file
path_PPO_MLP_std_2Mio = "PPO_MLP_std_2Mio/progress.csv" 
path_PPO_MLP_rew100_2Mio = "PPO_MLP_rew100_2Mio/progress.csv" 
path_baseline_rew100_2Mio = "baseline_rew100_2Mio/progress.csv" 
path_DQN_MLP_std_2Mio = "DQN_MLP_std_2Mio/progress.csv" 
path_DQN_MLP_rew100_2Mio = "DQN_MLP_rew100_2Mio/progress.csv" 
path_A2C_MLP_std_2Mio= "A2C_MLP_std_2Mio/progress.csv" 
path_A2C_MLP_rew100_2Mio= "A2C_MLP_rew100_2Mio/progress.csv" 


# Read the CSV file into a pandas DataFrame
df_PPO_MLP_std_2Mio = pd.read_csv(path_PPO_MLP_std_2Mio)
df_PPO_MLP_rew100_2Mio = pd.read_csv(path_PPO_MLP_rew100_2Mio)
df_baseline_rew100_2Mio = pd.read_csv(path_baseline_rew100_2Mio)
df_DQN_MLP_std_2Mio = pd.read_csv(path_DQN_MLP_std_2Mio)
df_DQN_MLP_rew100_2Mio = pd.read_csv(path_DQN_MLP_rew100_2Mio)
df_A2C_MLP_std_2Mio = pd.read_csv(path_A2C_MLP_std_2Mio)
df_A2C_MLP_rew100_2Mio = pd.read_csv(path_A2C_MLP_rew100_2Mio)
```

```python
import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8-darkgrid')  # This will apply a style similar to the provided image

# Define the colors
colors = {
    'df_PPO_MLP_std_2Mio': '#A0D2DB',  # light blue
    'df_PPO_MLP_rew100_2Mio': '#5B84B1',  # light green
    'df_baseline_rew100_2Mio': '#FF6F61',  # light red
    'df_DQN_MLP_rew100_2Mio': '#D2C2FF',  # light purple
    'df_A2C_MLP_rew100_2Mio': '#E7C8A0'  # light orange
}
s = 5
alpha=0.5

plt.figure(figsize=(10, 6))
# plt.scatter(df_PPO_MLP_std_2Mio['time/total_timesteps'], df_PPO_MLP_std_2Mio["rollout/ep_rew_mean"], alpha=alpha, label='df_PPO_MLP_std_2Mio', color=colors['df_PPO_MLP_std_2Mio'], s=s)
plt.scatter(df_PPO_MLP_rew100_2Mio['time/total_timesteps'], df_PPO_MLP_rew100_2Mio["rollout/ep_rew_mean"], alpha=alpha, label='df_PPO_MLP_rew100_2Mio', color=colors['df_PPO_MLP_rew100_2Mio'], s=s)
plt.scatter(df_baseline_rew100_2Mio['time/total_timesteps'], df_baseline_rew100_2Mio["rollout/ep_rew_mean"], alpha=alpha, label='df_baseline_rew100_2Mio', color=colors['df_baseline_rew100_2Mio'], s=s)
plt.scatter(df_DQN_MLP_rew100_2Mio['time/total_timesteps'], df_DQN_MLP_rew100_2Mio["rollout/ep_rew_mean"], alpha=alpha, label='df_DQN_MLP_rew100_2Mio', color=colors['df_DQN_MLP_rew100_2Mio'], s=s)
plt.scatter(df_A2C_MLP_rew100_2Mio['time/total_timesteps'], df_A2C_MLP_rew100_2Mio["rollout/ep_rew_mean"], alpha=alpha, label='df_A2C_MLP_rew100_2Mio', color=colors['df_A2C_MLP_rew100_2Mio'], s=s)
plt.title("rollout/ep_rew_mean over Time")
plt.xlabel('Total timestamps')
plt.ylabel("rollout/ep_rew_mean")
plt.legend()
plt.grid(True)
plt.show()
```
