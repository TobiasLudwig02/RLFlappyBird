```python
import pandas as pd
import os

log_dir = "./logs/"
df = pd.read_csv(os.path.join(log_dir, 'episode_rewards.csv'))
print(df.head(-10))
```

```python
# Plotten der Daten
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(df['timesteps'], df['reward'], label='Cumulative Reward per Episode')
plt.xlabel('Timesteps')
plt.ylabel('Cumulative Reward')
plt.title('Training Progress')
plt.legend()
plt.grid()
plt.show()
```

```python
# Scatterplot der Daten
plt.figure(figsize=(10, 5))
plt.scatter(df['timesteps'], df['reward'], label='Cumulative Reward per Episode')
plt.xlabel('Timesteps')
plt.ylabel('Cumulative Reward')
plt.title('Training Progress')
plt.legend()
plt.grid()
plt.show()
```
