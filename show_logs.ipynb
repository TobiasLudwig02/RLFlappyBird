```python
import pandas as pd
import os

# Load the CSV file
log_dir = "./logs/"  # Replace with your actual log directory
csv_file = log_dir + "progress.csv"  # Replace with the actual name of your CSV file if different

# Read the CSV file into a pandas DataFrame
df = pd.read_csv(csv_file)

print(df.head(-10))
```

```python
# Plotten der Daten
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(df['timesteps'], df['reward'], label='Cumulative Reward per Episode')
plt.xlabel('Timesteps')
plt.ylabel('Cumulative Reward')
plt.title('Training Progress')
plt.legend()
plt.grid()
plt.show()
```

```python
# Scatterplot der Daten
plt.figure(figsize=(10, 5))
plt.scatter(df['timesteps'], df['reward'], label='Cumulative Reward per Episode')
plt.xlabel('Timesteps')
plt.ylabel('Cumulative Reward')
plt.title('Training Progress')
plt.legend()
plt.grid()
plt.show()
```
